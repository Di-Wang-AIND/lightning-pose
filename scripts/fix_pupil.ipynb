```python
import pandas as pd
import os 
```

```python
data_dir = "/home/jovyan/litpose-data/ibl-pupil/"
file_name = 'CollectedData_full.csv'
df = pd.read_csv(os.path.join(data_dir, file_name), index_col=0, header=[0,1,2]) # load as is
df.head()
```

```python
df_pre = df.copy() # make a copy of the original data
print("before changes, nan count:")
print(df[("mic", "pupil_top_r", "x")].isnull().sum())
print(df[("mic", "pupil_bottom_r", "x")].isnull().sum())
print(df[("mic", "pupil_left_r", "y")].isnull().sum())
print(df[("mic", "pupil_right_r", "y")].isnull().sum())

mean_x_top_bot = (df[("mic", "pupil_top_r", "x")] + df[("mic", "pupil_bottom_r", "x")]) / 2.
mean_isnan = mean_x_top_bot.isnull()
df[("mic", "pupil_top_r", "x")].loc[~mean_isnan] = mean_x_top_bot.loc[~mean_isnan]
df[("mic", "pupil_bottom_r", "x")].loc[~mean_isnan] = mean_x_top_bot.loc[~mean_isnan]
# do the same for y in left and right 
mean_y_left_right = (df[("mic", "pupil_left_r", "y")] + df[("mic", "pupil_right_r", "y")]) / 2.
mean_y_is_nan = mean_y_left_right.isnull()
df[("mic", "pupil_left_r", "y")].loc[~mean_y_is_nan] = mean_y_left_right.loc[~mean_y_is_nan]
df[("mic", "pupil_right_r", "y")].loc[~mean_y_is_nan] = mean_y_left_right.loc[~mean_y_is_nan]

print("after changes, nan count:")
print(df[("mic", "pupil_top_r", "x")].isnull().sum())
print(df[("mic", "pupil_bottom_r", "x")].isnull().sum())
print(df[("mic", "pupil_left_r", "y")].isnull().sum())
print(df[("mic", "pupil_right_r", "y")].isnull().sum())
```

```python
fixed_filename = os.path.join(data_dir, file_name[:-4] + "_fixed.csv")
df.to_csv(fixed_filename)
```
